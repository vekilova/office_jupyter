```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tabl = pd.read_csv("datasets/office_episodes.csv", index_col = 0)
index = tabl.index
size_np = np.array(tabl.iloc[:,11])
sizelist = list()
markerlist = list()
for i in range(len(size_np)) :
    if size_np[i] == False :
        sizelist.append(25)
        markerlist.append('o')
    else :
        sizelist.append(250)
        markerlist.append('*')     
```

```python
colorlist = list()
color_np = np.array(tabl.iloc[:,12])
#print(color_np)
for i in range(len(color_np)) :
    if color_np[i] < 0.25 :
        colorlist.append("red")
    elif color_np[i] >= 0.25 and color_np[i] < 0.50 :
        colorlist.append("orange")
    elif color_np[i] >= 0.50 and color_np[i] < 0.75 :
        colorlist.append("lightgreen")
    else :
        colorlist.append("darkgreen")
#print(colorlist)
```

```python
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]
for j in range(len(color_np)) :
    plt.scatter(index[j], tabl.iloc[j,5], c = colorlist[j], s = sizelist[j], marker = markerlist[j])
plt.xlabel("Episode Number", size = 25)
plt.ylabel("Viewership (Millions)", size = 25)
plt.title("Popularity, Quality, and Guest Appearances on the Office", size = 25)

plt.show()
```
